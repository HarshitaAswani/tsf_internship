### **Data Modelling**

```python
# splitting the data into training and testing set
X =df.iloc[:, :-1].values
y =df.iloc[:, 1].values
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y,train_size=0.8,test_size=0.2,random_state=0)
```

```python
# fitting linear regression into training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# predicting the test set result
y_predict = regressor.predict(X_test)
y_predict
```

```python
# calculating the coefficient
coefficient = regressor.coef_
coefficient
```

```python
# calculating the intercept
intercept = regressor.intercept_
intercept
```

```python
# calculating the R squared value
from sklearn.metrics import r2_score
r2_score(y_test, y_predict)
```

```python
# predicting the score when the student studies for 9.25 hours a day
score = regressor.predict([[9.25]])
print('Score of a student is', score, 'when studies for 9.25 hours a day')
```
