```python
from sklearn.cluster import KMeans
```

```python
# elbow method
df1=df.iloc[:, :4]
sum_squared_error=[]
kclusters=range(1,15)
for k in kclusters:
    kmean=KMeans(n_clusters=k).fit(df1)
    sum_squared_error.append(kmean.inertia_)
```

```python
# elbow method
plt.plot(kclusters, sum_squared_error)
plt.xlabel('k')
plt.ylabel('Sum Squared Error')
plt.title('Predicting the optimum number of clusters')
plt.show()
```

```python
# k=3 is the optimum number of clusters
kmean=KMeans(n_clusters=3)
kmean.fit(df1)
```

```python
# fitting and predicting
y_predict=kmean.fit_predict(df1)
y_predict
```

```python
# centroid of cluster
kmean.cluster_centers_
```

```python
# checking the unique categories of Species column
df['Species'].value_counts()
```

```python
# converting df1 into an array
df2=np.array(df1)
df2
```

```python
# plotting the clusters
plt.scatter(df2[y_predict==1,2], df2[y_predict==1,3], color='g', label='Iris-setosa')
plt.scatter(df2[y_predict==2,2], df2[y_predict==2,3], color='b', label='Iris-versicolor')
plt.scatter(df2[y_predict==0,2], df2[y_predict==0,3], color='r', label='Iris-virginica')
plt.legend()
plt.show()
```
